```python
import networkx as nx
H = nx.DiGraph()
#adding nodes
H.add_nodes_from([
    (0, {"color": "pink", "size": 250}),
    (1, {"color": "red", "size": 400}),
    (2, {"color": "yellow", "size": 800}),
    (3, {"color": "orange", "size": 600}),
    (4, {"color": "gray", "size": 300})
    
])

#adding edges
H.add_edges_from([(0, 1),(1, 2),(1, 0),(1, 3),(2, 3),(3, 0),(1,4),(0,4),(4,2),(3,4),(4,4)])
node_colors = nx.get_node_attributes(H, "color").values()
colors = list(node_colors)
node_sizes = nx.get_node_attributes(H, "size").values()
sizes = list(node_sizes)
import matplotlib.pyplot as plt

#Plotting Graph

nx.draw_networkx(H, with_labels=True, node_color=colors, node_size=sizes)
plt.title("Directed Graph Neural Network")

plt.show()
```

```python
#converting to undirected graph
G = H.to_undirected()
nx.draw_networkx(G, with_labels=True, node_color=colors, node_size=sizes)
plt.title("UnDirected Graph Neural Network")
plt.show();
```
