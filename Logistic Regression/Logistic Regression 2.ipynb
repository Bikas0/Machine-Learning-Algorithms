```python
import pandas as pd
```

```python
dataset = pd.read_csv("data.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.Company.unique()
```

```python
dataset.Company.value_counts()
```

```python
dataset.columns
```

```python
for dollar in ['Close/Last', 'Open', 'High', 'Low']:
    dataset[dollar] = dataset[dollar].str.replace('$', '')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
# Convert the 'column_name' column from string to float
for column_name in ['Close/Last', 'Open', 'High', 'Low']:
    dataset[column_name] = dataset[column_name].astype(float)
```

```python
dataset.info()
```

```python
dataset.describe
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop("Date", axis = 1)
```

```python
import seaborn as sns
```

```python
sns.pairplot(dataset)
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
dataset["Company"]= label_encoder.fit_transform(dataset["Company"])

dataset["Company"].unique()
```

```python
X = dataset.drop("Company", axis = 1)
X
```

```python
Y = dataset["Company"]
```

```python
Y
```

```python
import matplotlib.pyplot as plt
plt.scatter(X["High"], Y)
```

```python
sns.distplot(X, bins = 10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state= 101)
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    "penalty":['l2', "l1"],
    "tol":[0.0001, 0.0002],
    "C":[1.0],
    "solver":['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    "max_iter":[100,200,300,400],
    "multi_class":['auto', 'ovr', 'multinomial'],
    "n_jobs":[1,2,3]
    
}
```

```python
grid = GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
print(grid.best_params_)
```
